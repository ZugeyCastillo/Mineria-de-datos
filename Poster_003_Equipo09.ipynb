### 3) Gráficas

```python
import matplotlib as plt
import matplotlib.pyplot as mpp
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import folium
```

**Gráfica 1**

```python
subdf['PdDistrict'].value_counts().sort_index().plot.bar()
```

En esta grpafica observamos el comportamiento de los delitos no resueltos en los barrios bajo estudio del condado de San Francisco, California. El principal barrio con mayor crímenes, o el más peligroso (por llamarlo de otra manera) es, sin duda, Southern.

**Gráfica 2**

```python
Distritos = subdf['PdDistrict'].value_counts()
plt.style.use('default')
cmap = plt.cm.summer
colors = cmap(np.linspace(0., 1., 10))
fig, ax = mpp.subplots(1, 2, figsize = (12,5), constrained_layout = True)
fig.suptitle('Ocurrencias por Distrito', fontsize = 16)
Distritos.plot.pie(autopct = "%.2f%%", colors = colors, ax = ax[1])
```

Distribución de delitos en el condado de San Francisco, California. Esta es otra manera de ver la Gráfica 1, pero en porcentajes, haciendola más entendible y atractiva visualmente.

**Gráfica 3**

```python
filt = (subdf['PdDistrict'] == 'SOUTHERN')
subdf.loc[filt, 'Category'].value_counts().sort_index().plot.bar()
```

Esta gráfica nos da una muestra de la distribución de delitos en el barrio con mayor crimen: Southern.

**Mapa 1**

```python
#https://www.kaggle.com/minhqvo/sanfrancisco-crime-analysis
t = subdf.PdDistrict.value_counts()

table = pd.DataFrame(data=t.values, index=t.index, columns = ['Cantidad'])
table = table.reset_index()
table.rename({'index': 'Barrios'}, axis='columns', inplace=True)

table
```

```python
gjson = r'https://cocl.us/sanfran_geojson'
sf_map = folium.Map(location = [37.77, -122.42], zoom_start = 12)
```

```python
sf_map.choropleth(
    geo_data=gjson,
    data=table,
    columns=['Barrios', 'Cantidad'],
    key_on='feature.properties.DISTRICT',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Barrios mas peligrosos de San Francisco'
)

sf_map
```

El mapa nos muestra el condado de San Francisco, California donde se diferencían los barrios por nivel de peligro según la cantidad de crímenes que suceden a diario. En el top 5 tenemos a:

1. Southern
1. Northern
1. Mission
1. Central
1. Bayview
