### Trabajando con bases de datos

```python
import pandas as pd
```

```python
df = pd.read_excel("Banco.xlsx")
df
```

Se tiene una base de datos de dimensión 7 columnas por 12 filas, donde se encuentra información personal de cada cliente que tiene el banco así como su monto en banco y estatus (1 o True para clientes activos, 0 o False para clientes inactivos). 

```python
df.columns
```

```python
df.dtypes
```

```python
df.isnull()
```

```python
df.isnull().any()
```

```python
df.isnull().sum()
```

Se observa que nuestra base de datos contiene valores nulos en las columnas "Estatus" y "Puntos", la función suma nos indica que son 1 y 3 valores, respectivamente.

```python
df = df.fillna(0)
df
```

```python
df.Nombre = df.Nombre.astype("str")
df.Correo = df.Correo.astype("str")
df.Telefono = df.Telefono.astype("int64")
df.Direccion = df.Direccion.astype("str")
df.Monto = df.Monto.astype("float")
df.Estatus = df.Estatus.astype("bool")
df.Puntos = df.Puntos.astype("int")
df
```

```python
subdf = pd.ExcelWriter('Banco1.xlsx', engine = 'xlsxwriter')
df.to_excel(subdf, index=False)
subdf.save()
```

### Tarea extra  
**Función astype()**  

```python
df = pd.read_excel("Banco.xlsx")
df['Puntos'] = df['Puntos'].fillna(999).astype(int)
df['Estatus'] = df['Estatus'].fillna(0.0).astype(float)
df
```

```python
# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dropna.html?highlight=dropna#pandas.DataFrame.dropna

df = pd.read_excel("Banco.xlsx")
df = df.dropna(subset=['Estatus'])
df
```
